### Part 1. Decision Tree Regression Theory/Intuition 

Imagine we have the following Data. We are given 2 inputs (X1 and X2) and our goal is to predict the output, the Y value. An example of this would be to predict someone's job based on their salary and years with the company, a problem we will encounter later on. First, lets look at the graph of the two inputs X1 and X2.


When it is scattered like this, linear and logistic regression would not be useful tools. Given that this is still a regression problem, we want to use another regression model. In this case, we'll use Decision Tree regression as our model. 

#### Here is what the decision tree regression will do to our data

The model makes split 1, then 2, etc. You may be wondering what the logic behind where the model makes these splits. Decision Tree Regression utilizes something called Informational Entropy. Informational Entropy asks whther or not a split will increase the amount of meaninful information about our points. It stops when there can be no additional information added in the process of splitting the data. This is a complex topic that you do not need to fully understand to use Decision Tree regression. If you are curious, see the wikipedia page and the khan academy video which explains it pretty well.

Wikipedia Link: https://en.wikipedia.org/wiki/Entropy_(information_theory)
Khan Academy Video: https://www.youtube.com/watch?v=2s3aJfRr9gE

#### Visualing our Decision Tree

We can visualize the graph as a tree of yes/no decisions, hence the name Decision Tree. Let's take a look.

#### Some Basic Tree Terminology
Each rectangle on the tree is a node. <br>
The node directly above any given node is a parent. <br>
The node directly under any given node is a child. <br>
The topmost node is called the root. <br>
Nodes without children are called leaves.

Let's say we have X1 = 30 and X3 = 50, which leaf will that coordinate end up in? <br> (Leaf 1 is the leftmost leaf and Leaf 5 is the rightmost leaf)

#### Making our prediction

How does the Decision Tree model make a prediction? It is quite simple actually. <br>It averages all the Y values of the data points in the leaf where the input coordinates lie.

So what would our model predict with X1 = 30 and X2 = 50 for inputs?

```python
#Import the Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading in the dataset
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

```python
#Creating our regresson
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X,y)
```

```python
#6.5 an arbitrary value for our prediction
y_pred = regressor.predict(6.5) 
```

```python
#Visualing the Decision Tree Regression results via MatPlotLib
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Job Title as a function of Position level and Salary')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
